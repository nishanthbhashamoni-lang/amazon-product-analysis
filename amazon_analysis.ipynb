```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('amazon.csv')
df.head()
```

```python
df.shape
df.columns
df.info()
df.isna().sum()
```

```python
df['discounted_price'] = df['discounted_price'].str.replace('₹','', regex=False)
df['discounted_price'] = df['discounted_price'].str.replace(',','', regex=False).astype(float)

df['actual_price'] = df['actual_price'].str.replace('₹','', regex=False)
df['actual_price'] = df['actual_price'].str.replace(',','', regex=False).astype(float)
```

```python
df['discount_percentage'] = df['discount_percentage'].str.replace('%','', regex=False).astype(float)
```

```python
df['rating_count'] = df['rating_count'].str.replace(',','', regex=False)
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')
```

```python
df['discount_amount'] = df['actual_price'] - df['discounted_price']
```

```python
df.info()
df.head()
```

```python
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
```

```python
df.info()
```

```python
df['rating_count'] = df['rating_count'].fillna(0)
```

```python
df['category'].value_counts().head(10)
```

```python
import matplotlib.pyplot as plt

df['category'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Product Categories")
plt.show()
```

```python
df.sort_values(by='rating', ascending=False)[
    ['product_name','category','rating','rating_count']
].head(10)
```

```python
df[['product_name','actual_price','discounted_price','discount_percentage']].sort_values(
    by='discount_percentage', ascending=False
).head(10)
```

```python
import seaborn as sns

sns.scatterplot(x='discount_percentage', y='rating', data=df)
plt.title("Discount vs Rating")
plt.show()
```

# Discount Analysis

```python
sns.scatterplot(x='discount_percentage', y='rating', data=df)
plt.title("Discount Percentage vs Rating")
plt.show()
```

## Key Insights

1. Electronics and accessories dominate the dataset.
2. Most products maintain ratings above 4.0.
3. High discounts do not always lead to better ratings.
4. A few products dominate the review counts, indicating strong popularity.

# Best Value Products

```python
df['value_score'] = (df['rating'] * df['discount_percentage']) / df['discounted_price']
```

```python
df.sort_values(by='value_score', ascending=False)[
    ['product_name','category','rating','discount_percentage','discounted_price']
].head(10)
```

# High Discount but Low Rating

```python
df[(df['discount_percentage'] > 50) & (df['rating'] < 3.5)][
    ['product_name','category','discount_percentage','rating']
]
```

# Category Wise Average Rating

```python
df.groupby('category')['rating'].mean().sort_values(ascending=False).head(10).plot(kind='bar')
plt.title("Top Categories by Average Rating")
plt.show()
```

# Most Popular Products

## Business Insights

1. Products with balanced discount and rating deliver the best value.
2. Some heavily discounted products still receive poor ratings, indicating quality concerns.
3. Certain categories dominate in terms of customer ratings.
4. Highly reviewed products represent strong customer trust and popularity.
